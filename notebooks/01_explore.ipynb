```python
import pandas as pd

orders = pd.read_csv('data/raw/olist_orders_dataset.csv')
order_items = pd.read_csv('data/raw/olist_order_items_dataset.csv')
order_reviews = pd.read_csv('data/raw/olist_order_reviews_dataset.csv')
customers = pd.read_csv('data/raw/olist_customers_dataset.csv')
products = pd.read_csv('data/raw/olist_products_dataset.csv')
sellers = pd.read_csv('data/raw/olist_sellers_dataset.csv')
category_translation = pd.read_csv('data/raw/product_category_name_translation.csv')

print("Before merge, order_items shape:", order_items.shape)
```

```python
# Convert to datetime so sorting actually works chronologically (not as text)
order_reviews['review_creation_date'] = pd.to_datetime(order_reviews['review_creation_date'])

# Sort by date, then keep only the latest review per order_id
order_reviews_dedup = order_reviews.sort_values('review_creation_date').drop_duplicates(subset='order_id', keep='last')

print("Original review rows:", order_reviews.shape[0])
print("After dedup:", order_reviews_dedup.shape[0])
print("Unique order_ids:", order_reviews['order_id'].nunique())
```

```python
# second merge
# Step 2: + reviews (deduplicated)
df = df.merge(order_reviews_dedup[['order_id', 'review_score']], on='order_id', how='left')
print("After merging reviews:", df.shape)

# Step 3: + customers
df = df.merge(customers, on='customer_id', how='left')
print("After merging customers:", df.shape)

# Step 4: + products
df = df.merge(products[['product_id', 'product_category_name']], on='product_id', how='left')
print("After merging products:", df.shape)

# Step 5: + sellers
df = df.merge(sellers, on='seller_id', how='left')
print("After merging sellers:", df.shape)

# Step 6: + category translation
df = df.merge(category_translation, on='product_category_name', how='left')
print("After merging category translation:", df.shape)

print()
print("Final columns:", df.columns.tolist())
```

```python
# Convert date columns to actual datetime objects first (they're currently just text/object dtype)
df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])
df['order_delivered_customer_date'] = pd.to_datetime(df['order_delivered_customer_date'])
df['order_estimated_delivery_date'] = pd.to_datetime(df['order_estimated_delivery_date'])

# How many rows have a missing delivered date? (order never delivered, or still in transit)
print("Missing delivered_customer_date:", df['order_delivered_customer_date'].isna().sum())
print("Total rows:", df.shape[0])
```

```python
# Recreate delay_flag using pandas' nullable boolean dtype, which properly supports NA
df['delay_flag'] = (df['order_delivered_customer_date'] > df['order_estimated_delivery_date']).astype('boolean')
df.loc[df['order_delivered_customer_date'].isna(), 'delay_flag'] = pd.NA

print(df['delay_flag'].dtype)
print(df['delay_flag'].value_counts(dropna=False))
```

```python
# For each seller, what fraction of their delivered orders were delayed?
# We use dropna() because delay_flag has NaN for undelivered orders — we only want
# to compute delay rate among orders we actually know the outcome for.

seller_delay = (
    df.dropna(subset=['delay_flag'])
      .groupby('seller_id')['delay_flag']
      .mean()
      .rename('seller_delay_rate')
)

print(seller_delay)
print()
print("Number of sellers:", seller_delay.shape[0])
```

```python
df = df.merge(seller_delay, on='seller_id', how='left')
print("After merging seller_delay_rate:", df.shape)
print(df[['seller_id', 'seller_delay_rate']].head(10))
```

```python
df
```

```python
# category_review_score: average review score per product category
category_review = (
    df.dropna(subset=['review_score'])
      .groupby('product_category_name_english')['review_score']
      .mean()
      .rename('category_review_score')
)

# seller_review_score: average review score per seller
seller_review = (
    df.dropna(subset=['review_score'])
      .groupby('seller_id')['review_score']
      .mean()
      .rename('seller_review_score')
)

# order_volume: how many order-items each seller has handled (a simple count, not a mean)
seller_volume = (
    df.groupby('seller_id')
      .size()
      .rename('order_volume')
)

# Merge all three back onto df
df = df.merge(category_review, on='product_category_name_english', how='left')
df = df.merge(seller_review, on='seller_id', how='left')
df = df.merge(seller_volume, on='seller_id', how='left')

print("Final shape:", df.shape)
print(df[['product_category_name_english', 'category_review_score', 
           'seller_id', 'seller_review_score', 'seller_delay_rate', 'order_volume']].head(10))
```
